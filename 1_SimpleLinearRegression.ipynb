```python
x = data["bmi"]
y = data["target"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)  # this is for the random split of the dataset on every run nut if i want to fix randomness it will randomly distributed only firse time then it is same for all time
#x_tarin,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1 0r 2 or anything )
```

training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
x_train
```

```python
y_train
```

```python
# here the model is in training...
model.fit(x_train.values.reshape(-1,1), y_train)
```

```python
model.coef_  # the coefficient of the equation of the linear regression y = Q0 + x*Q1
```

```python
model.intercept_ # the intercept of the the line of the quation of linear regression
```

```python
y_pred = model.predict(x_test.values.reshape(-1,1))
# model is predicting the target value....
```

```python
y_pred
```

```python
 plt.scatter(x_test,y_test,color = 'black', label = "Actual")
 plt.plot(x_test,y_pred,color = "blue",linewidth = 3 , label = "Predicted linear regression line")
```

```python
 plt.scatter(x_test,y_test,color = 'black', label = "Actual")
 plt.plot(x_test,y_pred,color = "blue",linewidth = 3 , label = "Predicted linear regression line")
 plt.xlabel("BMI")
 plt.ylabel("one year progression target ")
 plt.title("Linear regression on diabetes")
 plt.legend()
```

model evaluation

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(x_test,y_test)
```
